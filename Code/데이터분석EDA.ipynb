```python
import pandas as pd

# Define the years you want to process
years = [2013,2014,2015,2016,2017,2019,2021,2020,2022,2023,2024]

# Initialize an empty list to store the processed DataFrames
dfs = []

# Loop through each year, process the data, and append to the list
for year in years:
    # Construct the file path for the current year
    file_path = f'/content/drive/MyDrive/경정데이터분석/경정데이터/정렬데이터/{year}년 정렬된 출주표 및 순위.csv'

    # Read the CSV file for the current year
    df = pd.read_csv(file_path)

    # Filter out rows where 'rank1', 'rank2', or 'rank3' columns are not numeric
    df = df[pd.to_numeric(df['rank1'], errors='coerce').notnull()]
    df = df[pd.to_numeric(df['rank2'], errors='coerce').notnull()]
    df = df[pd.to_numeric(df['rank3'], errors='coerce').notnull()]

    # Convert 'rank1', 'rank2', and 'rank3' columns to integers
    df['rank1'] = df['rank1'].astype(int)
    df['rank2'] = df['rank2'].astype(int)
    df['rank3'] = df['rank3'].astype(int)

    # Append the processed DataFrame to the list
    dfs.append(df)

# Concatenate all DataFrames in the list into one DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

# Now `combined_df` contains the concatenated data from 2013 to 2022
```
