```python
import pandas as pd

# Define the years you want to process
years = [2013,2014,2015,2016,2017,2019,2021,2020,2022,2023]

# Initialize an empty list to store the processed DataFrames
dfs = []

# Loop through each year, process the data, and append to the list
for year in years:
    # Construct the file path for the current year
    file_path = f'/content/drive/MyDrive/경정데이터분석/경정데이터/정렬데이터/{year}년 정렬된 출주표 및 순위.csv'

    # Read the CSV file for the current year
    df = pd.read_csv(file_path)

    # Filter out rows where 'rank1', 'rank2', or 'rank3' columns are not numeric
    df = df[pd.to_numeric(df['rank1'], errors='coerce').notnull()]
    df = df[pd.to_numeric(df['rank2'], errors='coerce').notnull()]
    df = df[pd.to_numeric(df['rank3'], errors='coerce').notnull()]

    # Convert 'rank1', 'rank2', and 'rank3' columns to integers
    df['rank1'] = df['rank1'].astype(int)
    df['rank2'] = df['rank2'].astype(int)
    df['rank3'] = df['rank3'].astype(int)

    # Append the processed DataFrame to the list
    dfs.append(df)

# Concatenate all DataFrames in the list into one DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

# Now `combined_df` contains the concatenated data from 2013 to 2022
```

```python
# 새로운 열을 추가하는 함수 정의
def create_rank_vector(row):
    # 기본 벡터 [0, 0, 0]
    rank_vector = [0, 0, 0]

    # 순서 값이 rank1, rank2, rank3 중 어디에 있는지 확인
    if row['순서'] == row['rank1']:
        rank_vector = [1, 0, 0]
    elif row['순서'] == row['rank2']:
        rank_vector = [0, 1, 0]
    elif row['순서'] == row['rank3']:
        rank_vector = [0, 0, 1]

    return rank_vector

# 데이터프레임에 새로운 'rank' 열 추가
combined_df['rank'] = combined_df.apply(create_rank_vector, axis=1)
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
class RankingDataset(Dataset):
    def __init__(self, features, labels):
        self.features = torch.tensor(features.values, dtype=torch.float32)  # Numpy 배열로 변환된 입력
        self.labels = torch.tensor(labels, dtype=torch.long)  # 레이블을 정수형으로 변환

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        return self.features[idx], self.labels[idx]
```

```python
# 1순위로 후처리
probability_re = probability_array.reshape(-1,6,4)
top3_indices_sorted = np.argsort(probability_re[:, :, 1], axis=1)[:, -3:][:, ::-1]
top3_indices_sorted = top3_indices_sorted+1
```

```python
# 1로 시작하는 인덱스를 사용하기 위해 top3_indices_sorted는 그대로 사용
top3_indices_sorted = top3_indices_sorted  # 이미 +1을 한 상태이므로 그대로 사용

# 0-based 인덱스를 사용하여 확률 값을 추출해야 하므로 다시 1을 빼서 0-based로 만듦
top3_indices_zero_based = top3_indices_sorted - 1

# 상위 3개의 확률 값을 추출
top3_probabilities = np.take_along_axis(probability_re[:, :, 1], top3_indices_zero_based, axis=1)

# 순위와 확률 값을 결합하여 출력
rank_and_probabilities = np.stack((top3_indices_sorted, top3_probabilities), axis=2)

# 출력 예시
for sample in rank_and_probabilities:
    print(sample)
```

```python
# 단승식
matches = []
matched_probabilities = []

# 확률 값이 일치한 경우의 평균을 계산하기 위한 수정된 코드
for rank_prob, row2 in zip(rank_and_probabilities, result):
    # rank_prob은 순위와 확률 값이 포함된 배열, row2는 비교 대상 배열
    match = False

    # 첫 번째 순위와 비교
    rank, prob = rank_prob[0]  # rank_prob에서 첫 번째 순위와 확률 추출
    if rank == row2[0]:  # row2의 첫 번째 값과 비교
        match = True
        matched_probabilities.append(prob)  # 일치한 경우 확률 값을 저장

    matches.append(match)

# 정확도 계산
accuracy = np.mean(matches)  # 적중한 경우의 비율

# 일치한 확률 값의 평균 계산
if matched_probabilities:
    average_matched_probability = np.mean(matched_probabilities)
else:
    average_matched_probability = 0  # 일치한 것이 없으면 0

print("정확도:", accuracy)
print("일치한 확률 값의 평균:", average_matched_probability)
```

```python
# 연승식
match_first_rank = []  # 1순위 비교 결과를 저장할 리스트
match_second_rank = []  # 2순위 비교 결과를 저장할 리스트
matched_probabilities_first = []  # 1순위 매칭된 확률 값을 저장할 리스트
matched_probabilities_second = []  # 2순위 매칭된 확률 값을 저장할 리스트

for rank_prob, result_row in zip(rank_and_probabilities, result):
    # rank_prob은 순위와 확률 값이 포함된 배열, result_row는 비교 대상 배열
    top1_rank, top1_prob = rank_prob[0]  # 첫 번째 순위와 확률
    top2_rank, top2_prob = rank_prob[1]  # 두 번째 순위와 확률

    # 1순위 비교
    match_first = top1_rank in result_row[:2]  # top1_rank가 result의 1, 2순위 중에 있는지 확인
    match_first_rank.append(match_first)

    # 2순위 비교
    match_second = top2_rank in result_row[:2]  # top2_rank가 result의 1, 2순위 중에 있는지 확인
    match_second_rank.append(match_second)

    # 매칭된 경우의 확률 값을 저장
    if match_first:
        matched_probabilities_first.append(top1_prob)
    if match_second:
        matched_probabilities_second.append(top2_prob)

# 정확도 계산
accuracy_first = np.mean(match_first_rank)  # 1순위 매치 비율
accuracy_second = np.mean(match_second_rank)  # 2순위 매치 비율

# 1순위와 2순위의 일치한 확률 값의 평균 계산
if matched_probabilities_first:
    average_matched_probability_first = np.mean(matched_probabilities_first)
else:
    average_matched_probability_first = 0  # 일치한 것이 없으면 0

if matched_probabilities_second:
    average_matched_probability_second = np.mean(matched_probabilities_second)
else:
    average_matched_probability_second = 0  # 일치한 것이 없으면 0

print("1순위 정확도:", accuracy_first)
print("2순위 정확도:", accuracy_second)
print("1순위 일치한 확률 값의 평균:", average_matched_probability_first)
print("2순위 일치한 확률 값의 평균:", average_matched_probability_second)
```
